```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("New_student_score.csv")
```

```python
#Gender Distribution
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

from the above chart, We have analysed that the number of female is more than number of male in this data.

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot = True)
plt.show()
```

From the above chart, We have conclude that the parent education is also affect to their children score.

```python
gb1 = df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True)
plt.show()
```

From the above chart we have analysed that weekly study hours can help to achieve good score.

```python
gb2 = df.groupby("NrSiblings").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb2
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb2, annot = True)
plt.show()
```

```python
#Distribution of EthnicGroup
df1 = df.drop("Unnamed: 0",axis=1)
groupA = df1.loc[(df["EthnicGroup"]=="group A")].count()
groupB = df1.loc[(df["EthnicGroup"]=="group B")].count()
groupC = df1.loc[(df["EthnicGroup"]=="group C")].count()
groupD = df1.loc[(df["EthnicGroup"]=="group D")].count()
groupE = df1.loc[(df["EthnicGroup"]=="group E")].count()

label = ["groupA","groupB","groupC","groupD","groupE"]
my_list = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
my_list
plt.pie(my_list,labels=label,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
